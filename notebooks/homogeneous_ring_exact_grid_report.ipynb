# Homogeneous Ring Exact Grid Report

This notebook does not launch the validation run. It reads the existing `results/homogeneous_ring_exact_grid` artifacts and displays the generated plots.

```python
import json
from pathlib import Path

from IPython.display import Image, display


def find_repo_root(start=None):
    if start is None:
        start = Path.cwd()
    for candidate in (start, *start.parents):
        if (candidate / "src" / "hrdmc").is_dir() and (candidate / "experiments").is_dir():
            return candidate
    raise RuntimeError("Run this notebook from inside the hardrod-qmc repository")


REPO_ROOT = find_repo_root()
RESULT_DIR = REPO_ROOT / "results" / "homogeneous_ring_exact_grid"
SUMMARY_PATH = RESULT_DIR / "summary.json"
print(REPO_ROOT)
print(SUMMARY_PATH)
```

```python
if not SUMMARY_PATH.exists():
    raise FileNotFoundError(
        "Missing summary.json. Run experiments/14_homogeneous_ring_exact_grid.py first."
    )

summary = json.loads(SUMMARY_PATH.read_text())
print("status:", summary["status"])
print("case_count:", summary["case_count"])
print("formula:", summary["formula"])
print("max abs error:", summary["max_energy_per_particle_abs_error"])
print("max relative error:", summary["max_energy_per_particle_relative_error"])
```

```python
plot_paths = [RESULT_DIR / item for item in summary.get("plots", [])]
if not plot_paths:
    plot_paths = sorted((RESULT_DIR / "plots").glob("*.png"))

if not plot_paths:
    raise FileNotFoundError("No PNG plots found under results/homogeneous_ring_exact_grid/plots")

for path in plot_paths:
    print(path.name)
    display(Image(filename=str(path)))
```

```python
failed = [row for row in summary["cases"] if not row["passed"]]
if failed:
    print("FAILED CASES")
    for row in failed:
        print(row["case_id"], row["max_energy_per_particle_abs_error"])
else:
    print("All cases passed the configured exact-energy tolerance.")
```
